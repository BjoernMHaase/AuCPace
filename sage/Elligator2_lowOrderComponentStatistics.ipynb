```python
def Curve25519PointFromXCoordinate(xCoordinate):
    prime = 2^255 - 19
    F = GF(prime)
    A = F(486662)
    nonsquare = F(2)   
    E = EllipticCurve(F, [0, A , 0, 1 , 0])
    Twist = EllipticCurve(F, [0, A * nonsquare, 0, 1 * nonsquare^2, 0])
    u = F(xCoordinate)
    d = 1
    v2 = u^3 + A*u^2 + u
    isOnTwist = 0
    if not v2.is_square():
        isOnTwist = 1
        #print("Input point is on the twist! "),
        E = Twist
        d = nonsquare
        u = d * u
        v2 = u^3 + A*u^2 * nonsquare + u * nonsquare^2
    v = v2.sqrt()
    
    point = E(u, v)
    return point, isOnTwist 

def elligator2_curve25519(r):
    p = 2^255 - 19
    F = GF(p)
    A = F(486662)
    B = F(1)
    r = F(r)
    
    u = F(2) # Our Non-square
    powerForChi = 2^254 - 10
    
    v = - A / (1 + u * r^2)
    epsilon = (v^3 + A * v^2 + B * v)^powerForChi
    x = epsilon * v - (1 - epsilon) * A/2
    return x

def isOnTheImageOf_elligator2(point):
    p = 2^255 - 19
    F = GF(p)
    A = F(486662)
    B = F(1)
    isOnImage = True
    x = point[0]
    y = point[1]
    
    u = F(2) # Our Non-square
    if (x == -A):
        isOnImage = False
    if ((y == 0) and (x != 0)):
        isOnImage = False
    if (isOnImage):
        testNumber = - u * x * (x + A);
        isOnImage = testNumber.is_square()
        
    return isOnImage

lowOrderPoints = []
OrderPrimeSubgroup = 2^252 + 27742317777372353535851937790883648493

for n in range(40):
    point, isOnTwist = Curve25519PointFromXCoordinate(n)
    if (not isOnTwist):
        point = point * (OrderPrimeSubgroup)
        existsAlready = point in lowOrderPoints
        if ((not isOnTwist) and (not existsAlready)):
            lowOrderPoints.append(point)
    if (len(lowOrderPoints) == 8):
        break;
        
for pt in lowOrderPoints:
    print(pt.order())
    print (pt)
    print("")

frequency = [0] * len(lowOrderPoints)

pF = 2^255 - 19
F = GF(pF)
FG = GF(OrderPrimeSubgroup)


for n in range(10):
    xCoordinate = elligator2_curve25519(F.random_element())
    point, isOnTwist = Curve25519PointFromXCoordinate(xCoordinate)
    lowOrderComponent = point * (OrderPrimeSubgroup)
    indexOfLowOrderPoint = lowOrderPoints.index(lowOrderComponent)
    frequency[indexOfLowOrderPoint] += 1

    
xCoordinate = elligator2_curve25519(1234)
point, isOnTwist = Curve25519PointFromXCoordinate(xCoordinate)
basePoint = point * 8;

numDuplicates = [0] * (len(lowOrderPoints) + 1)
    
for n in range(1000):
    randomExponent = FG.random_element()
    point = basePoint * Integer(randomExponent)
    numDuplicatesOnImage = 0
    for LO in lowOrderPoints:
        isOnImage = isOnTheImageOf_elligator2 (point + LO)
        if isOnImage:
            numDuplicatesOnImage += 1
                
    numDuplicates[numDuplicatesOnImage] += 1
    print ("Add ", numDuplicatesOnImage," :", numDuplicates)
    
print (numDuplicates)
#print (frequency)
```
